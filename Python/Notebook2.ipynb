```python
import pandas as pd
```

```python
file = "C:/Users/Oscar/Documents/Proyecto_Data_Science/sales_train.csv"
```

```python
df = pd.read_csv(file,nrows=100000) 
df["date"]=pd.to_datetime(df["date"])
df["Month"]=df.date.dt.to_period("M")
```

```python
df2=df.groupby(["Month","shop_id","item_id"])["item_cnt_day"].sum().reset_index()
```

```python
shop18=df2[(df2["shop_id"]==18)&(df2["item_id"]==22092)]
shop18["Month"] = shop18["Month"].astype(str)
new_index = list(map(lambda x: '2013-%02d' % x,range(1,13)))
shop18 = shop18.set_index("Month").reindex(new_index, fill_value=0)
shop18
```

```python
shop18=shop18.drop(columns=["shop_id","item_id"])
```

```python
shop18.rolling(window=3,min_periods=1).mean()
```

```python
shop18["maverage"]=shop18.rolling(window=3,min_periods=1).mean()
```
